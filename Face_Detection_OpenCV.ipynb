```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
img = plt.imread("./group-photo.jpg")
```

```
plt.imshow(img)
```

```
# haarcascade library
detector = cv2.CascadeClassifier("./haarcascade_frontalface_alt.xml")
```

```
#detecting
faces = detector.detectMultiScale(img)
```

```
faces
```

```
faces.shape
```

```
faces[0]
```

```
x, y, w, h = faces[0]
```

```
x
```

```
y
```

```
w
```

```
h
```

```
# rectangle(image, coordinate of top left, coordinate of bottom right , RGB color of rectangle, thickness of rectangle)
img = cv2.rectangle(img, (x, y) , (x+w , y+h), (255,255,255), 3)
```

```
plt.imshow(img)
```

Plot rect for all faces

```
for face in faces:
  x,y,w,h = face
  img = cv2.rectangle(img, (x, y) , (x+w , y+h), (255,255,255), 3)
```

```
plt.imshow(img)
```
